```python
name = input("Enter student name: ")
marks= input("Enter marks: ")
print("Student Name: ", name)
print("Marks: ", marks)
if(marks >= '85'):
    print("Grade: A")
elif(marks >= '75'):
    print("Grade: B")
elif(marks >= '65'):
    print("Grade: C")
elif(marks >= '55'):
    print("Grade: D")
else:
    print("Grade: F")
```
